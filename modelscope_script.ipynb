```python
from modelscope.hub.api import HubApi

api = HubApi()
owner = "OpenMOSS"
page = 1
context = []

models = api.list_models(owner_or_group=owner, page_size=50, page_number=page)
models["Models"][0]
```

```python
while True:
    try:
        # 调用 SDK 方法获取模型列表
        models_data = api.list_models(
            owner_or_group=owner,
            page_size=50,
            page_number=page
        )

        # 如果没有更多模型，退出循环
        if not models_data:
            break

        # 遍历当前页的模型
        for model in models_data:
            likes = model.get("Likes", 0)
            if likes >= 50:
                context.append({
                    "name": model.get("Name"),
                    "description": model.get("Summary"),  # 简介
                    "created_at": model.get("CreatedTime"),
                    "updated_at": model.get("UpdatedTime"),
                    "stars": likes,
                    "url": f"https://www.modelscope.cn/models/{model['Owner']}/{model['Name']}"
                })

        # 如果当前页模型数不足 page_size，说明是最后一页
        if len(models_data) < 50:
            break

        page += 1

    except Exception as e:
        print(f"获取模型时出错：{e}")
        break

# 输出结果
for item in context:
    print(f"[{item['stars']}❤️] {item['name']}: {item['description']} - {item['url']}")
```
